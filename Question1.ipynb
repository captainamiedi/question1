```python
import pandas as pb
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
url='https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv'
```

```python
data = pb.read_csv(url, header=0)
```

```python
data.head()
```

```python
data.shape
```

```python
fig, ax = plt.subplots()
ax.scatter(data['lifeExp'], data['gdpPercap'])
ax.set_title('Data')
ax.set_xlabel('life exp')
ax.set_ylabel('gdp')
```

```python
data['year']
```

```python
newYear =data[data.year == 1987]
```

```python
newYear.head()
```

```python
fig, ax = plt.subplots()
ax.scatter(newYear['gdpPercap'], newYear['lifeExp'], marker='D', alpha=0.5)
ax.set_title('Data')
ax.set_xlabel('life Rxp')
ax.set_ylabel('gdp Per Cap')
```

```python
y1 = newYear[newYear['continent'] == 'Africa']
```

```python
y1.head()
```

```python
y2 = newYear[newYear['continent'] == 'Asia']
```

```python
y3 = newYear[newYear['continent'] == 'Europe']
```

```python
y4 = newYear[newYear['continent'] == 'Americas']
y5 = newYear[newYear['continent'] == 'Oceania']
```

```python
plt.scatter(newYear,y1,color='blue', marker= '*', label='Africa')
plt.scatter(newYear,y2,color= 'red', marker='v', label='Asia')
plt.scatter(newYear,y3,color= 'green', marker='.', label='Europe')
plt.scatter(newYear,y4,color= 'pink', marker='.', label='Americas')
plt.scatter(newYear,y5,color= 'yellow', marker='.', label='Oceania')
```
